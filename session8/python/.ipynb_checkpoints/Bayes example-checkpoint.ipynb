```python
import numpy as np
import scipy as sp
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
N = 100

data = np.abs(2*np.random.randn(N,2) + np.array([4,5]))

fig1 = plt.figure(figsize=(10,4))
plt.scatter(data[:,0], data[:,1], color='blue', s=10)
plt.xlim(0, 10)
plt.ylim(0, 10)


# plt.savefig('data.svg')
```

```python
mu_1 = np.array([2.5, 5])
mu_2 = np.array([5, 5])
mu_3 = np.array([7.5, 5])

Sigma_true = 100*np.eye(2)
Sigma_viz = np.eye(2)
```

```python
fig2 = plt.figure(figsize=(10,4))
ax2 = fig2.add_subplot(111)
ax2.scatter(data[:,0], data[:,1], color='blue', s=10)
plt.xlim(0, 10)
plt.ylim(0, 10)

for mu in [mu_1, mu_2, mu_3]:

    rv = multivariate_normal(mean=mu, cov=Sigma_true)
    rv_viz = multivariate_normal(mean=mu, cov=Sigma_viz)

    x, y = np.mgrid[0:10:.01, 0:10:.01]
    pos = np.dstack((x, y))
    
    ax2.contour(x, y, rv_viz.pdf(pos))
    

# plt.savefig('data_with_likelihood.svg')
```

```python
log_likelihood = np.zeros(3)

for i, mu in enumerate([mu_1, mu_2, mu_3]):
    log_likelihood[i] = np.sum(multivariate_normal.logpdf(data, mean=mu, cov=Sigma_true))

print(log_likelihood)
```

```python
# z=1 : student house, z=2: grandma Jane, z=3 family with kids

prior = np.array([0.3, 0.1, 0.6])
log_prior = np.log(prior)
log_joint = log_likelihood + log_prior

print(log_prior)
print(log_likelihood)
print(log_joint)
```

```python
log_marginal = np.log(np.sum(np.exp(log_joint)))
log_posterior = log_joint - log_marginal
posterior = np.exp(log_posterior)
print(posterior)
```

```python
plt.figure(figsize=(10,4))
plt.stem([1,2,3], prior)
plt.stem([1,2,3], posterior, linefmt='C1--', markerfmt='C1s')
plt.xticks([1,2,3])
plt.xlabel('z', fontsize=18)
plt.ylabel('probability', fontsize=18)
plt.legend(['prior', 'posterior'], loc='upper center', fontsize=18)

# plt.savefig('posterior.svg')
```

### Decision


```python
loss_z_zhat = np.array([[0,10,1], [1,0,1], [1,10,0]])

expected_posterior_loss = np.sum(loss_z_zhat*posterior[:, np.newaxis], axis=0)
plt.figure(figsize=(10,4))
plt.stem([1,2,3], expected_posterior_loss)
plt.xticks([1,2,3])
plt.xlabel('z', fontsize=18)
plt.title('posterior expected loss', fontsize=18)

# plt.savefig('expect_posterior_loss.svg')
```

### Posterior predictive

$$
\begin{aligned}
p(\mathbf{x}_{\text{new}} | \mathbf{x}_1, ..., \mathbf{x}_N) &= \sum_{i=1}^3 p(\mathbf{x}_{\text{new}}, z=i | \mathbf{x}_1, ..., \mathbf{x}_N) \\ 
&= \sum_{i=1}^3 p(\mathbf{x}_{\text{new}} | z=i) p(z=i | \mathbf{x}_1, ..., \mathbf{x}_N)\\ 
&= \sum_{i=1}^3 \mathcal{N}(\mathbf{x}_{\text{new}}; \boldsymbol{\mu}_i, \sigma^2 \mathbf{I}) p(z=i | \mathbf{x}_1, ..., \mathbf{x}_N)
\end{aligned}
$$


```python
# Posterior predictive

x, y = np.mgrid[0:10:.01, 0:10:.01]
pos = np.dstack((x, y))

pred_1 = multivariate_normal(mean=mu_1, cov=Sigma_viz).pdf(pos)
pred_2 = multivariate_normal(mean=mu_2, cov=Sigma_viz).pdf(pos)
pred_3 = multivariate_normal(mean=mu_3, cov=Sigma_viz).pdf(pos)

post_pred = pred_1*posterior[0] + pred_2*posterior[1] + pred_3*posterior[2]


fig3 = plt.figure(figsize=(10,4))
ax3 = fig3.add_subplot(111)
ax3.scatter(data[:,0], data[:,1], color='blue', s=10)
plt.xlim(0, 10)
plt.ylim(0, 10)

ax3.contour(x, y, post_pred, 10)

# plt.savefig('predictive posterior.svg')
```

### MAP prediction

The likelihood where we plug the MAP estimate of the latent variable.

$$
\begin{aligned}
p(\mathbf{x}_{\text{new}} | z=\hat{z}^{\text{MAP}}) = \mathcal{N}(\mathbf{x}_{\text{new}}; \boldsymbol{\mu}_1, \sigma^2 \mathbf{I})
\end{aligned}
$$


```python
 # MAP prediction

x, y = np.mgrid[0:10:.01, 0:10:.01]
pos = np.dstack((x, y))

pred_MAP = multivariate_normal(mean=mu_1, cov=Sigma_viz).pdf(pos)

fig4 = plt.figure(figsize=(10,4))
ax4 = fig4.add_subplot(111)
ax4.scatter(data[:,0], data[:,1], color='blue', s=10)
plt.xlim(0, 10)
plt.ylim(0, 10)

ax4.contour(x, y, pred_MAP, 10)

# plt.savefig('MAP_prediction.svg')
```

$$
