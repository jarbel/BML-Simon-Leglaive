```python
import numpy as np
import scipy as sp
from scipy.stats import multinomial
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
N = 100

data = np.abs(2*np.random.randn(N,2) + np.array([4,5]))

fig1 = plt.figure(figsize=(10,4))
plt.scatter(data[:,0], data[:,1], color='blue', s=10)
plt.xlim(0, 10)
plt.ylim(0, 10)


plt.savefig('data.svg')
```

```python
mu_1 = np.array([2.5, 5])
mu_2 = np.array([5, 5])
mu_3 = np.array([7.5, 5])

z = np.array([1, 2, 3])
```

```python
fig2 = plt.figure(figsize=(10,4))
ax2 = fig2.add_subplot(111)
ax2.scatter(data[:,0], data[:,1], color='blue', s=10)
plt.xlim(0, 10)
plt.ylim(0, 10)

for mu in [mu_1, mu_2, mu_3]:

    rv = multivariate_normal(mean=mu, cov=np.eye(2))

    x, y = np.mgrid[0:10:.01, 0:10:.01]
    pos = np.dstack((x, y))
    
    ax2.contour(x, y, rv.pdf(pos))
    

plt.savefig('data_with_likelihood.svg')
```

```python
log_likelihood = np.zeros(3)

for i, mu in enumerate([mu_1, mu_2, mu_3]):
    log_likelihood[i] = np.sum(multivariate_normal.logpdf(data, mean=mu, cov=np.eye(2)))

print(log_likelihood)
```

```python
# z=1 : student house, z=2: grandma Jane, z=3 family with kids

prior = np.array([30, 10, 60])/100
log_prior = np.log(prior)
log_joint = np.zeros(3)
indiv_marginal = 0

for i, mu in enumerate([mu_1, mu_2, mu_3]):
    log_like = multivariate_normal.logpdf(data, mean=mu, cov=np.eye(2))
    log_joint[i] = np.sum(log_like + log_prior[i])
    indiv_marginal += np.exp(log_like + log_prior[i])
    
    
log_marginal = np.sum(np.log(indiv_marginal))
log_posterior = log_joint - log_marginal

print(log_posterior)
```
