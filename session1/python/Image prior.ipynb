```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
from scipy.stats import norm
```

```python
im = np.array(Image.open('cameraman.png'))/255*2 - 1

plt.figure(figsize=(7,7))
plt.imshow(im, cmap='gray')
plt.colorbar()

# plt.savefig('cameraman.svg')
```

```python
grad_x = np.diff(im)
grad_x_vec = grad_x.flatten()

plt.figure(figsize=(7,7))
plt.imshow(grad_x, cmap='gray')
plt.colorbar()

# plt.savefig('cameraman_grad.svg')
```

```python
(mean, std) = norm.fit(grad_x_vec)
```

```python
fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(111)

x = np.linspace(-1.5, 1.5, 100)
ax.plot(x, norm.pdf(x, loc=mean, scale=std), 'k-', lw=2, alpha=0.6, label='Gaussian log-pdf')
_ = ax.hist(grad_x_vec, density=True, histtype='stepfilled', alpha=0.5, bins=100, label='histogram')

handles, labels = ax.get_legend_handles_labels()
# sort both labels and handles by labels
labels, handles = zip(*sorted(zip(labels, handles), key=lambda t: t[0]))
ax.legend(handles[::-1], labels[::-1], fontsize=14)

# plt.savefig('cameraman_gaussian_linear.svg')
```

```python
fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(111)

x = np.linspace(-1.5, 1.5, 100)
ax.plot(x, norm.pdf(x, loc=mean, scale=std), 'k-', lw=2, alpha=0.6, label='Gaussian log-pdf')
_ = ax.hist(grad_x_vec, density=True, histtype='stepfilled', alpha=0.5, bins=100, label='histogram')

handles, labels = ax.get_legend_handles_labels()
# sort both labels and handles by labels
labels, handles = zip(*sorted(zip(labels, handles), key=lambda t: t[0]))
ax.legend(handles[::-1], labels[::-1], fontsize=14)

plt.yscale('log')
plt.ylim([0.01, 100])

# plt.savefig('cameraman_gaussian_log.svg')
```

```python
(shape, loc, scale) = t.fit(grad_x_vec)

fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(111)

x = np.linspace(-1.5, 1.5, 100)
ax.plot(x, t.pdf(x, df=shape, loc=loc, scale=scale), 'r--', lw=2, alpha=0.6, label='Student''s t log-pdf')
ax.plot(x, norm.pdf(x, loc=mean, scale=std), 'k-', lw=2, alpha=0.6, label='Gaussian log-pdf')
_ = ax.hist(grad_x_vec, density=True, histtype='stepfilled', alpha=0.5, bins=100, label='histogram')

handles, labels = ax.get_legend_handles_labels()
# sort both labels and handles by labels
labels, handles = zip(*sorted(zip(labels, handles), key=lambda t: t[0]))
ax.legend(handles[::-1], labels[::-1], fontsize=14)

plt.yscale('log')
plt.ylim([0.01, 100])

# plt.savefig('cameraman_gaussian_student_log.svg')
```
